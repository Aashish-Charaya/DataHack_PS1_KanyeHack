```python

# Define the file paths of your Excel files
file_paths = ['unicorn cb 11 dec 15.xlsx', 'unicorn cb 23 dec 16.xlsx', 'unicorn cb 23 dec 17.xlsx','unicorn cb 29 dec 18.xlsx','unicorn cb 31 dec 19.xlsx','unicorn cb 31 dec 20.xlsx','unicorn cb 31 dec 21.xlsx']  # Add the paths of your files

# Read each Excel file into a DataFrame
dfs = [pd.read_excel(file) for file in file_paths]

# Concatenate the DataFrames
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged DataFrame to a new Excel file
merged_df.to_excel('merged_file.xlsx', index=False)
```

```python
merged_df = merged_df[(merged_df["Country"] == "India") & (merged_df["Industry"] == "Fintech")]
merged_df.shape
```

```python
# Convert 'Joined' column to date format
merged_df['Date Joined'] = pd.to_datetime(merged_df['Date Joined'])

# Add 3 years to each year's data
merged_df['Date Joined'] = merged_df['Date Joined'] + pd.DateOffset(years=3)
```

```python
merged_df['Valuation ($B)'] = merged_df['Valuation ($B)'].str.replace('$', '').astype(float)
merged_df
```

```python
aggregated_df = merged_df.groupby('Date Joined')['Valuation ($B)'].sum()

# Reset the index to convert 'Date Joined' back to a column
aggregated_df = aggregated_df.reset_index()

print(aggregated_df)
```
